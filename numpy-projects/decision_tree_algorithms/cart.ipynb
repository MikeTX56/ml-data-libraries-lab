#### CART (Classification and Regression Tree)

its cos algorithm specifically for classification it uses `Gini impurity` to recursively find binary splts that maximize class purity in child nodes.

 it selects the best split by minimizing the weighted sum impurity of the tworesulting subsets.

```python
import numpy as np 
```

##### Gini Impurity
$$G = 1 - \sum_{i=1}^{C} (p_i)^2$$


```python
def gini_impurity(y):
    if len(y) == 0:
        return 0
    p = np.unique(y, return_counts=True)[1]/len(y)
    return 1 - np.sum(p**2)
```

##### Regression cost: Mean Squared Error (MSE)
$$MSE = \frac{1}{m} \sum_{i=1}^{m} (y_i - \bar{y})^2$$


```python
def mse_cost(y):
    if len(y) == 0:
        return 0
    return np.mean((y - np.mean(y))**2)
```

##### Weighted Split Cost
$$J(k, t_k) = \frac{m_{left}}{m} G_{left} + \frac{m_{right}}{m} G_{right}$$


##### For clasification problem

```python
def cart_cost(left_y, right_y):
    m = len(left_y) + len(right_y)
    if m == 0:
        return 0
    
    m_left = len(left_y)/m
    m_right = len(right_y)/m

    return m_left * gini_impurity(left_y) + m_right *gini_impurity(right_y)
```

##### For regression problem
The `CART` algorithm works mostly the same way as earlier, except that instead of try‐
ing to split the training set in a way that minimizes impurity, it now tries to split the
training set in a way that minimizes the `MSE`.

$$J(k, t_k) = \frac{m_{left}}{m} MSE_{left} + \frac{m_{right}}{m} MSE_{right}$$


```python
def cart_cost_reg(left_y, right_y):
    m = len(left_y) + len(right_y)
    if m == 0:
        return 0
    m_left = len(left_y)/m
    m_right = len(right_y)/m

    return m_left * mse_cost(left_y) + m_right * mse_cost(right_y)
```
